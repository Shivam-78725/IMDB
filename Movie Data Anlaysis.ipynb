```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
def rating(rating):
    if rating >= 7.0:
        return 'Excellent'
    elif rating >= 5:
        return 'Average'
    else:
        return "Don't Watch it!"
```
